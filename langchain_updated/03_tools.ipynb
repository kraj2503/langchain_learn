```python
from langchain.chat_models import init_chat_model


model = init_chat_model(model='gpt-4.1-nano')
```

```python
from langchain.tools import tool

@tool ## <- Decorator which will make a tool
def get_weather(city:str) -> str:
    """Get the weather of the city""" ## Doc string  
    
    return f"weather in {city} is rainy"

model_with_tools = model.bind_tools([get_weather])
```

```python
response = model_with_tools.invoke("Whats the weather in pune?")

for tool_call in response.tool_calls:
    print(f"Tool: {tool_call['name']}")
    print(f"Args: {tool_call['args']}")
    
print(response)
```

```python
messages = [{"role":"user","content":"Whats the weather in Pune?"}]

ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)


for tool_call in ai_msg.tool_calls:
    print(tool_call)
    tool_res = get_weather.invoke(tool_call)
    messages.append(tool_res)
print(f"\n\n\n {messages} \n\n\n")
final_res = model_with_tools.invoke(messages)
print(final_res.text)
```
